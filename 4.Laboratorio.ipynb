EJERCICIO # 1

¿Cuál es el tamaño en bytes de un array de 1000 elementos de tipo booleano?

```python
import numpy as np

# Tamaño en bytes de un array de 1000 elementos booleano
booleanos = np.zeros(1000, dtype=bool)
bytes = booleanos.nbytes
print(f"Tamaño en bytes de un array de 1000 elementos booleanos: {bytes} bytes")
```

EJERCICIO #4 

Dados los arrays a = [1, 4, 2, 7] y b = [1, 3, 2, 9], obtenga la media aritmética de la diferencia (a-b).

```python
#Media aritmética de la diferencia (a - b)
a = np.array([1, 4, 2, 7])
b = np.array([1, 3, 2, 9])
media_de_la_diferencia = np.mean(a - b)
print(f"Media aritmética de la diferencia (a - b): {media_de_la_diferencia}")
```

EJERCICIO #5

Tengo valores de cordenadas (x, y) en las columnas del array
1.33 4.5
30.0 10.7
70.2 0.5

Agregue a este array las coordenadas (37.1, -3.6). Muestre en pantalla las dimensiones del nuevo array.

```python
# Agregar coordenadas al array y mostrar sus dimensiones
coordenadas = np.array([[1.33, 4.5],
                        [30.0, 10.7],
                        [70.2, 0.5]])

nuevas_coordenadas = np.array([[37.1, -3.6]])
coordenadas = np.vstack((coordenadas, nuevas_coordenadas))
print("Dimensiones del nuevo array de coordenadas:")
print(coordenadas.shape)
```
